```python
qpath = r'../../../../../' # path to quetzal here
data = r'inputs/'
import sys
sys.path.append(qpath + r'quetzal/')

# import class
from quetzal.io.gtfs_reader.feed_gtfsk import Feed
```

### Read GTFS

```python
feed = Feed(path=data + r'accra_trotro', dist_units='m')
feed.describe()
```

### Map GTFS

```python
feed.map_stops(feed.stops['stop_id'])
```

```python
feed.map_routes(
    feed.routes['route_id'].head(10)
)
```

```python
feed.map_trips(
    feed.trips.groupby('route_id').first().trip_id.head(50)
)
```

### Export as GIS

```python
feed.write_geojson(
    'outputs/accra/' # folder name
)
```
